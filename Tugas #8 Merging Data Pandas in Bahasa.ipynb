# Tugas #8 Merging Data Pandas in Bahasa

• Pada Soal kedua ini kalian akan menggunakan Data Video Youtube yang Trending di US tahun 2017-2018. Data tersebut dapat kamu akses melalui google classroom dengan nama “Usvideos.csv"

• Apa yang harus kamu lakukan? Ikuti step berikut :
1. Buat dataframe dengan nama 'title' yang berisi video_id dan title.
2. Buat dataframe dengan nama 'description' yang berisi video_id dan description.
3. Export keduanya ke dalam CSV baru dan import Kembali menggunakan glob.
4. Lakukan operasi merge ( boleh dengan .merge atau .join) kepada dataframe title dan description. Kemudian munculkan title video yang tidak ada descriptionnya.
5. Ada berapa nilai unik title yang tidak ada descriptionnya?

```python
usvideos = pd.read_csv('USvideos.csv')
usvideos
```

```python
# 1. Buat dataframe dengan nama 'title' yang berisi video_id dan title.
title = usvideos[['video_id', 'title']]
title
```

```python
# 2. Buat dataframe dengan nama 'description' yang berisi video_id dan description.
description = usvideos[['video_id', 'description']]
description
```

```python
# 3.a. Export keduanya ke dalam CSV baru
title.to_csv('USvideos_' + 'title' + '.csv', index=False)
description.to_csv('USvideos_' + 'description' + '.csv', index=False)
```

```python
# 3.b. Import Kembali menggunakan glob.
from glob import glob
file_name = glob('USvideos_*')
print(file_name)

df_title_desc = []
for i in file_name:
    df = pd.read_csv(i)
    df_title_desc.append(df)
```

```python
# 3.b. Import Kembali menggunakan glob. (description)
df_title_desc[0].head()
```

```python
# 3.b. Import Kembali menggunakan glob. (title)
df_title_desc[1].head()
```

```python
# 4. Lakukan operasi merge (boleh dengan .merge atau .join) kepada dataframe title dan description. Kemudian munculkan title video yang tidak ada descriptionnya.
all_data = pd.merge(title, description, how='outer')
all_data[all_data['description'].isnull()]
```

```python
# 5. Ada berapa nilai unik title yang tidak ada descriptionnya?
unik_title = len(all_data[all_data['description'].isnull()].drop_duplicates())
print(f'Nilai unik title yang tidak ada decriptionnya sebanyak {unik_title}')
```
